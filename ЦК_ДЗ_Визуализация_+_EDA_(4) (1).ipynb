Анализ графика в Задании 1 (Диаграмма разброса Vmag vs e_Plx)

Что представляют эти переменные?

**Vmag** – Видимая величина звезды (яркость звезды, чем выше значение, тем менее яркая звезда).

**e_Plx** – Погрешность измерения параллакса (чем больше значение, тем менее точное измерение расстояния до звезды).

**Что показывает диаграмма разброса?**

На графике каждая точка представляет собой одну звезду, а ее координаты – это значения Vmag и e_Plx.

**Что можно сказать по графику?**

Разброс точек – Звезды с разными Vmag (яркостью) имеют разные уровни погрешности (e_Plx).

Тренды:

* Если Vmag увеличивается (звезды тусклее), то e_Plx становится больше. Это логично, так как более тусклые звезды находятся дальше, и их параллакс измерять сложнее.
* Если Vmag низкий (звезды яркие), то e_Plx обычно низкий, что указывает на более точные измерения параллакса для ярких звезд.
* Возможные выбросы – Если на графике есть точки с очень высоким e_Plx, это может указывать на звезды с большой ошибкой измерения расстояния.

Вывод:

Этот график показывает, что чем тусклее звезда (Vmag высокий), тем сложнее измерить ее расстояние (выше e_Plx). Это соответствует ожиданиям астрономов, поскольку далекие объекты труднее измерять с высокой точностью. Также можно проверить, есть ли выбросы, которые могут повлиять на анализ.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Указываем путь к файлу в Google Colab
file_path = "/content/sample_data/train_var_1.csv"

# Загружаем данные
df = pd.read_csv(file_path)

# Проверка данных
print(df.head())  # Вывод первых строк
print(df.info())  # Информация о данных

# Преобразуем числовые столбцы в float (на случай, если они загружены как object)
num_cols = ["Vmag", "Plx", "e_Plx", "B-V", "Amag"]
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Фильтруем только числовые столбцы
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Построение boxplot
fig, axes = plt.subplots(len(num_cols), 1, figsize=(8, 5 * len(num_cols)))

for i, col in enumerate(num_cols):
    sns.boxplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f"Boxplot для {col}")
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()
```

```python
# Выбираем числовые переменные
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Создаем фигуру и несколько субплотов (один график с несколькими частями)
fig, axes = plt.subplots(len(num_cols), 1, figsize=(10, 5 * len(num_cols)))

# Построение boxplot для каждой переменной в отдельном subplot
for i, col in enumerate(num_cols):
    sns.boxplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f"Boxplot для {col}", fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].grid(True)

plt.tight_layout()  # Делаем графики аккуратными
plt.show()
```

**Каждый ящик на графике показывает распределение данных для одной переменной (например, Vmag, Plx).**

💡 **Элементы графика:**

* **Ящик (box)** – показывает центральные 50% данных (межквартильный размах, IQR).
* **Линия в середине (медиана)**– это "середина" данных (значение, которое делит их пополам).
* **"Усы" (whiskers)**– показывают диапазон большинства данных.
* **Точки за пределами усов** – это выбросы (аномально большие или маленькие значения).




📌**Общий вывод по boxplot:**

🔹 Boxplot (ящики с усами) для всех числовых переменных
Этот график позволяет визуально увидеть распределение значений и выбросы для каждой числовой переменной:

* Vmag (видимая величина звезды) Большинство значений сгруппированы в середине, но есть выбросы.
Некоторые звезды значительно ярче или тусклее, чем большинство.

* Plx (параллакс, измерение расстояния)
Наличие выбросов – возможно, есть аномально близкие звезды или ошибки измерений.

* e_Plx (погрешность измерения параллакса)
Есть звезды с большой погрешностью измерений, что может быть связано с трудностями в наблюдениях.
* B-V (цветовой индекс звезды)
Большинство звезд имеют цветовой индекс в типичном диапазоне.

* Amag (абсолютная звездная величина, истинная яркость)

Значительный разброс значений, что логично, так как звезды могут быть как карликами, так и гигантами.

📌 Вывод: В данных есть выбросы, особенно в Plx и e_Plx. Это может быть связано с погрешностями измерений или редкими звездами.

```python
# 🔹 Задание 6: Графики парных зависимостей (возьмем часть переменных)
selected_cols = ["Vmag", "Plx", "e_Plx", "Amag"]  # Выбрали основные числовые переменные
sns.pairplot(df[selected_cols], diag_kind="hist")
plt.show()
```

Этот график строит гистограммы распределений (диагональ) и диаграммы рассеяния (остальные клетки) для числовых переменных.

Диагональ (гистограммы): Показывает, как распределена каждая переменная.
Остальные графики: Показывают зависимость одной переменной от другой (насколько они связаны).


🔍 Разбираем зависимости между переменными

1️⃣ **Vmag (видимая звездная величина) vs Amag (абсолютная звездная величина)**
Видна четкая линейная зависимость – чем больше Vmag, тем больше Amag.

Это связано с тем, что абсолютная и видимая величина измеряют яркость, но в разных масштабах.
Вывод: Эти параметры очень похожи, возможно, один из них можно исключить при моделировании.

2️⃣ **Plx (параллакс) vs Vmag**
Наблюдается слабая отрицательная связь:
Маленький Plx (звезды далеко) → Vmag (яркость) выше (тусклые звезды).
Большой Plx (звезды близко) → Vmag меньше (яркие звезды).

Далекие звезды обычно тусклее, а близкие – ярче.
Вывод: Подтверждает, что Plx помогает определять, насколько далеко и насколько тусклая звезда.

3️⃣ **Plx vs e_Plx (погрешность измерения параллакса)**
Нет четкой зависимости.

Погрешность измерения (e_Plx) распределена случайно и не связана напрямую с Plx.
Вывод: Погрешность в измерении расстояния может быть высокой даже у близких звезд.

4️⃣ **Amag vs Plx**
Есть разброс: звезды с одинаковой абсолютной яркостью могут быть на разном расстоянии.

Разные классы звезд (карлики, гиганты) могут иметь одинаковую яркость, но находиться на разном расстоянии.
Вывод: Нельзя судить о расстоянии звезды только по Amag.
